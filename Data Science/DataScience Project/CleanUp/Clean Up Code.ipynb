```python
# Step 1. Load the `titanic_data.csv` file into a pandas dataframe
RawO = pd.read_table('Raw_Outputs.csv', sep=',',index_col='Ser No.')
RawO
```

```python
CompTab = RawO.iloc[:, :10].copy()
CompTab

# # Assuming new_data has the same structure (11 columns) as first_11_columns
# combined_data = pd.concat([first_11_columns, new_data], ignore_index=True)
```

```python
BCCTab = [col for col in RawO.columns if 'BCC' in col]
BCCTab = RawO[BCCTab].copy()
BCCTab
```

```python
# Find columns with "Vol" in the header
vol_bcc = [col for col in BCCTab.columns if 'Vol' in col]

# Sum the values in the "Vol" columns and store the result in a new column
BCCTab['Tot Vol Frac BCC'] = BCCTab[vol_bcc].sum(axis=1)

# Print the DataFrame with the new column
BCCTab
```

```python
FCCTab = [col for col in RawO.columns if 'FCC' in col]
FCCTab = RawO[FCCTab].copy()
FCCTab
```

```python
# Find columns with "Vol" in the header
vol_fcc = [col for col in FCCTab.columns if 'Vol' in col]

# Sum the values in the "Vol" columns and store the result in a new column
FCCTab['Tot Vol Frac FCC'] = FCCTab[vol_fcc].sum(axis=1)

# Print the DataFrame with the new column
FCCTab
```

```python
# Get the columns with "Tot" from the respective DataFrames
BCCTotCols = [col for col in BCCTab.columns if 'Tot' in col]
FCCTotCols = [col for col in FCCTab.columns if 'Tot' in col]

# Select the columns from the DataFrames
BCCTotDF = BCCTab[BCCTotCols]
FCCTotDF = FCCTab[FCCTotCols]

# Concatenate the DataFrames horizontally (axis=1)
BCC_FCC = pd.concat([BCCTotDF, FCCTotDF], axis=1)

# Print the new DataFrame
BCC_FCC
```

```python
# Find columns with "Vol" in the header
vol_both = [col for col in BCC_FCC.columns if 'Vol' in col]
volDF = pd.DataFrame()
volDF['Tot Vol Frac BCC+FCC'] = BCC_FCC[vol_both].sum(axis=1)
# volDF

# Concatenate the DataFrames horizontally (axis=1)
All_FCC_BCC = pd.concat([BCC_FCC, volDF], axis=1)
All_FCC_BCC
```

```python
# # Subtract BCC+FCC from 1
volODF=pd.DataFrame()
volODF['Tot Vol Frac Other']=1-All_FCC_BCC['Tot Vol Frac BCC+FCC']
volODF

#combine tables
All_Vols = pd.concat([All_FCC_BCC,volODF], axis=1)
All_Vols
```

```python
Comp_Vols = pd.concat([CompTab,All_Vols], axis=1)

Comp_Vols
```
